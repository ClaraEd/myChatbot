```python
s = {'Which_situation': survey.iloc[:,3]}
s1 = {'Which_situation': survey.iloc[:,7]}
s2 = {'Which_situation': survey.iloc[:,11]}

f = {'feeling': survey.iloc[:,4]}
f1 = {'feeling': survey.iloc[:,8]}
f2 = {'feeling': survey.iloc[:,12]}

d = {'do': survey.iloc[:,5]}
d1 = {'do': survey.iloc[:,9]}
d2 = {'do': survey.iloc[:,13]}

a = {'after': survey.iloc[:,6]}
a1 = {'after': survey.iloc[:,10]}
a2 = {'after': survey.iloc[:,14]}
```

```python
s2 = pd.DataFrame(s2).dropna().reset_index(drop=True)
s1 = pd.DataFrame(s1).dropna().reset_index(drop=True)
s = pd.DataFrame(s).dropna().reset_index(drop=True)

f2 = pd.DataFrame(f2).dropna().reset_index(drop=True)
f1 = pd.DataFrame(f1).dropna().reset_index(drop=True)
f = pd.DataFrame(f).dropna().reset_index(drop=True)

d2 = pd.DataFrame(d2).dropna().reset_index(drop=True)
d1 = pd.DataFrame(d1).dropna().reset_index(drop=True)
d = pd.DataFrame(d).dropna().reset_index(drop=True)

a2 = pd.DataFrame(a2).dropna().reset_index(drop=True)
a1 = pd.DataFrame(a1).dropna().reset_index(drop=True)
a = pd.DataFrame(a).dropna().reset_index(drop=True)
```

```python
# # # df['Which_situation']=intents_sit+intents_sit2 + intents_sit3
# frames = [s+s1]

situation = pd.concat([s,s1,s2], join='outer')
feelings = pd.concat([f,f1,f2], join='outer')
doing = pd.concat([d,d1,d2], join='outer')
afterwards = pd.concat([a,a1,a2], join='outer')
```

```python
situation['situation'] = np.repeat('situation',len(situation))

feelings['feelings'] = np.repeat('feelings',len(feelings))

doing['doing'] = np.repeat('doing',len(doing))

afterwards['afterwards'] = np.repeat('afterwards',len(afterwards))
```

```python
situation.to_csv('situation_intents.csv', index=False)
feelings.to_csv('feelings_intents.csv', index=False)
doing.to_csv('doing_intents.csv', index=False)
afterwards.to_csv('afterwards_intents.csv', index=False)
```
